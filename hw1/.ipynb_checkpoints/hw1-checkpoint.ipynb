```julia
using Convex
using ECOS
```

```julia
m = 100
n = 30

A = randn(m,n)
b = randn(m,1)

x1 = Variable(n)
x2 = Variable(n)
x3 = Variable(n)
x4 = Variable(n)
```

```julia
# First figure
problem1 = minimize(norm(A*x1-b,1))
solve!(problem1, ECOSSolver())

# Second figure
problem2 = minimize(norm(A*x2-b,2))
solve!(problem2, ECOSSolver())

# Third figure
problem3 = minimize(sum(max(0, abs(A*x3-b) - 0.5)))
solve!(problem3, ECOSSolver())

# Forth figure
problem4 = minimize(sum(-log(1-b+A*x4)-log(1+b-A*x4)), b - A*x4 < 1, b - A*x4 > -1)
solve!(problem4, ECOSSolver())
```

```julia
# First figure

using PyPlot

figure(figsize=(20,5))
r = -2:0.1:2
plt[:hist](b-A*x1.value, 100, facecolor = "white")
plot(r, 10*abs(r), "k")
axis([-2,2,0,40])
```

```julia
# Second figure

using PyPlot

figure(figsize=(20,5))
r = -2:0.1:2
plt[:hist](b-A*x2.value, 100, facecolor = "white")
plot(r, abs(r).^2, "k")
axis([-2,2,0,10])
```

```julia
# Third figure

using PyPlot

figure(figsize=(20,5))
r = -2:0.1:2
plt[:hist](b-A*x3.value, 100, facecolor = "white")
plot(r, 20*max(0, abs(r) - 0.5), "k")
axis([-2,2,0,40])
```

```julia
using PyPlot

figure(figsize=(20,5))
r = -1:0.0001:1
plt[:hist](0.5*b-A*x4.value, 50, facecolor = "white")
plot(r, -log(1-r.^2), "k")
axis([-2,2,0,10])
```
